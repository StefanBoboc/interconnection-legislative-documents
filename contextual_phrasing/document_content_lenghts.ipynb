```python
#
# Create a sorted list of the documents to be processed. 
# The list is ordered from the document with the fewest mentions to the document with the most mentions.
# List is dumped in doc_id_len_sorted.json
#

import mysql.connector as mysql
import pandas as pd

mydb = mysql.connect(
    host="127.0.0.1",
    user="root",
    passwd="xxxxx",
    # database="licentade10"
    database="preprocessed"
)

query = f"SELECT * FROM unique_document"
df_document = pd.read_sql(query, con=mydb)

if df_document.empty:
    print("DataFrame is empty")
    exit(1)

doc_id_list = []
doc_len_list = []
for index, row in df_document.iterrows():
    # Read file with the content of the document
    with open(row['doc_content'], 'r', encoding='utf-8') as file:
        doc_content = file.read()

    doc_id_list.append(row['doc_id'])
    doc_len_list.append(len(doc_content))
```

```python
# Get the sorted indices
sorted_indices = sorted(range(len(doc_len_list)), key=lambda k: doc_len_list[k])

# Use the sorted indices to reorder both doc_len_sort and doc_id_sort
doc_len_sort = [doc_len_list[i] for i in sorted_indices]
doc_id_sort = [doc_id_list[i] for i in sorted_indices]
```

```python
dict2 = dict(zip(doc_id_sort, doc_len_sort))
```

```python
dict2
```

```python
import json
with open('doc_id_len_sorted.json', 'w') as file:
    json.dump(doc_id_sort, file)
```
